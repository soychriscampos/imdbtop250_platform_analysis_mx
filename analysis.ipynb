## Key Questions:

## Distribution Analysis
### Which are the common genres?

```python
# Gender separation:
genre_separation = df["Género"].str.split(", ").explode()

# Frequency
genre_count = genre_separation.value_counts()

genre_count.head(5).plot(kind="bar", figsize=(10,5),
                                         title = "Top Géneros de las mejores 250 películas de IMBDb")
plt.show()
```

### Rating Distribution

```python
df["Calificación"].hist(bins=20, figsize=(10,5))
plt.title("Distribución de las mejores 250 películas de IMBDb")
plt.xlabel("Calificación")
plt.ylabel("Cantidad de Películas")
plt.show()
```

### Cinema Evolution

```python
df["Año"].value_counts().sort_index().plot(kind="line", figsize=(12,6), 
                                           title="Cantidad de Películas por Año")
plt.show()
```

## Streaming Platform Analysis

## Which streaming platform includes the highest number movies from IMDb's top 250

```python
platforms = df["Plataforma"].str.split(", ").explode()
platforms_count = platforms.value_counts()
platforms_count.plot(kind="bar", figsize=(10,5), 
                     title="¿Qué plataforma de streaming en México tiene más películas del top 250 de IMDb?")
#plt.xticks(rotation=45)
plt.show()
```

### How many movies aren't in "Ninguna Plataforma"

```python
df[df["Plataforma"] == "Ninguna Plataforma"].shape[0]
```

### Which movies are more popular (rating) en each platform?

```python
streaming_platforms = df[df["Plataforma"] != "Ninguna Plataforma"].copy()
streaming_platforms["Plataforma"] = streaming_platforms["Plataforma"].str.split(", ").explode().reset_index(drop=True)

platforms_avg_rating = streaming_platforms.groupby("Plataforma")["Calificación"].mean().sort_values(ascending=False)

round(platforms_avg_rating, 1)
```

```python
plt.figure(figsize=(10,5))
platforms_avg_rating.sort_values(ascending=False).plot(kind="bar",
                                                      title="Calificación promedio de Películas por Plataforma")
plt.ylim(8, 9)
plt.ylabel("Calificación IMDb")
plt.xticks(rotation=45)
plt.show()
```
